1. Graphical Method

Example 1: Linear Fit (y=mx+c)

Task: Fit a straight line to the data points.

Data:

    x: [1, 2, 3, 4, 5]
    y: [2, 4, 5, 4, 5]

```python
import matplotlib.pyplot as plt
import numpy as np

# Data points
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

# Fit a straight line (linear regression)
coefficients = np.polyfit(x, y, 1)  # 1 indicates a linear fit
m, c = coefficients

# Line equation: y = mx + c
fit_line = m * x + c

# Plot
plt.scatter(x, y, color='red', label='Data Points')  # Scatter plot for data
plt.plot(x, fit_line, label=f'Best Fit: y = {m:.2f}x + {c:.2f}', color='blue')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.title('Graphical Method: Linear Fit')
plt.grid()
plt.show()
```

1. Input Data: The data points are given as arrays for x and y.
2. Best Fit Line: Using np.polyfit, we compute the slope (m) and intercept (c) of the best-fit line.
3. Visualization: The scatter plot shows actual data points, while the line represents the model's prediction.

2. Method of Least Squares

Example 2: Quadratic Fit (y=a + bx + cx^2)

Task: Fit a parabola to the data points.

Data:

    x: [0, 1, 2, 3, 4]
    y: [1, 1.8, 1.3, 2.5, 6.3]

```python
# Quadratic Fit
coefficients = np.polyfit(x, y, 2)  # 2 indicates a quadratic fit
a, b, c = coefficients

# Parabola equation: y = ax^2 + bx + c
fit_curve = a * x**2 + b * x + c

# Plot
plt.scatter(x, y, color='green', label='Data Points')
plt.plot(x, fit_curve, label=f'Best Fit: y = {a:.2f}x² + {b:.2f}x + {c:.2f}', color='purple')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.title('Method of Least Squares: Quadratic Fit')
plt.grid()
plt.show()
```

1. Model Selection: Here, a quadratic model is chosen.
2. Parameter Estimation: Using np.polyfit, we estimate the coefficients a,b,c.
3. Result Analysis: The fit is shown as a smooth curve that minimizes the squared differences between actual and predicted values.

3. Fitting Curves

Example 3: Exponential Fit (y= a * e^(bx) )

Task: Fit an exponential curve to data points.

Data:

    x: [0, 1, 2, 3, 4]
    y: [2.5, 3.5, 7.4, 20.5, 54.6]

```python
from scipy.optimize import curve_fit

# Exponential function
def exponential_func(x, a, b):
    return a * np.exp(b * x)

# Fit the data
params, _ = curve_fit(exponential_func, x, y)
a, b = params

# Predicted values
fit_curve = exponential_func(x, a, b)

# Plot
plt.scatter(x, y, color='orange', label='Data Points')
plt.plot(x, fit_curve, label=f'Best Fit: y = {a:.2f}e^({b:.2f}x)', color='red')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.title('Curve Fitting: Exponential Fit')
plt.grid()
plt.show()
```

1. Non-Linear Model: We fit y= a * e^(bx), where a and b are constants.
2. Curve Fit: Using curve_fit from scipy, the parameters are optimized to minimize errors.
3. Graph: The curve demonstrates the relationship between x and y following an exponential growth pattern.

4. Method of Group Averages

Example 4: Grouped Data Linear Fit (y=a+bx)

Task: Group data into two subsets and calculate the line of best fit.

Data:

    Group 1: x: [1, 2, 3], y: [2, 4, 6]
    Group 2: x: [4, 5, 6], y: [8, 10, 12]

```python
# Averages of groups
x1, y1 = np.mean([1, 2, 3]), np.mean([2, 4, 6])
x2, y2 = np.mean([4, 5, 6]), np.mean([8, 10, 12])

# Calculate slope and intercept
b = (y2 - y1) / (x2 - x1)
a = y1 - b * x1

# Best fit line
x_full = np.linspace(1, 6, 100)
y_full = a + b * x_full

# Plot
plt.scatter([1, 2, 3, 4, 5, 6], [2, 4, 6, 8, 10, 12], color='brown', label='Data Points')
plt.plot(x_full, y_full, label=f'Best Fit: y = {a:.2f} + {b:.2f}x', color='blue')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.title('Method of Group Averages: Linear Fit')
plt.grid()
plt.show()
```

1. Group Division: Data is divided into two groups, and averages are calculated.
2. Line Calculation: The slope b and intercept a are determined from group averages.
3. Validation: The grouped method approximates the overall trend.

5. Laws Containing Three Constants

In this section, we focus on fitting models with three constants. For example, equations of the form y= a * e^(bx) + c or y=a + bx + c*ln⁡(x).

Example 5: Fitting y=a * e^(bx) + c

Task: Fit the data using an exponential model with an added constant c.

Data:

    x: [0, 1, 2, 3, 4]
    y: [2.7, 5.4, 10.8, 19.5, 40.2]

```python
from scipy.optimize import curve_fit
import numpy as np
import matplotlib.pyplot as plt

# Data points
x = np.array([0, 1, 2, 3, 4])
y = np.array([2.7, 5.4, 10.8, 19.5, 40.2])

# Define the model function
def three_const_func(x, a, b, c):
    return a * np.exp(b * x) + c

# Provide initial guesses for the parameters (a, b, c)
initial_guesses = [1, 0.5, 1]

# Fit the data with increased maxfev and initial guesses
params, _ = curve_fit(three_const_func, x, y, p0=initial_guesses, maxfev=5000)
a, b, c = params

# Predicted values
fit_curve = three_const_func(x, a, b, c)

# Plot
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x, fit_curve, label=f'Best Fit: y = {a:.2f}e^({b:.2f}x) + {c:.2f}', color='green')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.title('Three Constants Model: Exponential with Offset')
plt.grid()
plt.show()
```

1. Logarithmic Term: This model accounts for logarithmic growth in the data, using the term cln⁡(x).
2. Parameter Estimation: Coefficients a,b,c are computed to minimize the sum of squared errors.
3. Validation: The fit curve highlights how well the model describes data trends.
4. p0=initial_guesses: Sets initial parameter guesses (a=1,b=0.5,c=1). You can adjust these based on your understanding of the data's behavior.
5. maxfev=5000: Increases the maximum number of function evaluations to ensure convergence.
6. Normalization (Optional): If the data range is wide, scaling x and y can help the optimizer.
