2. Bisection Method

Example 1: Screenshot 2024-12-09 at 12-44-07 Сomputational Mathematics.png

```python
def bisection_method(f, a, b, tol):
    if f(a) * f(b) >= 0:
        print("Invalid initial values. f(a) and f(b) must be of different signs.")
        return None

    midpoint = (a + b) / 2 
    while abs(f(midpoint)) > tol:
        if f(a) * f(midpoint) < 0: 
            b = midpoint 
        else:
            a = midpoint 
        midpoint = (a + b) / 2 
    return midpoint

# Definition of a function
def f(x):
    return x**3 - 4*x - 9

# Using bisection method
root = bisection_method(f, 2, 3, 1e-6)
print(f"Approximate root: {root}")
```

1. The bisection method checks for a change in the sign of the function at the ends of the segment [a,b].
2. If f(a) and f(b) have different signs, the root lies between them.
3. The iterations continue until the value of the function in the middle of the segment becomes less than the specified accuracy tol.

Example 2: f(x)=cos⁡(x)−x

```python
def f(x):
    return np.cos(x) - x

# Using bisection method
root = bisection_method(f, 0, 1, 1e-6)
print(f"Approximate root: {root}")
```

1. The function f(x)=cos(x)−x has a root in the interval [0,1].
2. The method iteratively reduces the size of the interval until accuracy is achieved.

3. Newton-Raphson Method

Example 1: Screenshot 2024-12-09 at 12-52-33 Сomputational Mathematics.png

```python
def newton_raphson(f, df, x0, tol):
    x = x0
    while abs(f(x)) > tol:
        x = x - f(x) / df(x)  # Formula 
    return x

def f(x):
    return x**3 - 2*x - 5

def df(x):
    return 3*x**2 - 2

# Using method
root = newton_raphson(f, df, 2.5, 1e-6)
print(f"Root: {root}")
```

1. The Newton-Raphson method requires a function f(x) and its derivative f′(x).
2. Iterations continue until the value of the function at the current point becomes less than tol.
3. The method converges quickly with a good initial approximation.

4. Secant Method

Example 1: Screenshot 2024-12-09 at 12-57-38 Сomputational Mathematics.png

```python
def secant_method(f, x0, x1, tol):
    # Iterative approximation of the root by the secant method
    while abs(f(x1)) > tol:
        x_temp = x1 - f(x1) * (x1 - x0) / (f(x1) - f(x0))  # Formula
        x0, x1 = x1, x_temp  # Updating values
    return x1

def f(x):
    return x**3 - 4*x - 9

# Using method
root = secant_method(f, 2, 3, 1e-6)
print(f"Root: {root}")
```

1. The secant method uses two initial points x0 and x1​.
2. The secant formula constructs an approximation based on the line connecting the points (x0​,f(x0​)) and (x1​,f(x1​)).
3. The iterations continue until the value of the function f(x) at the current point becomes less than tol.

Example 2: f(x)=cos⁡(x)−x

```python
def f(x):
    return np.cos(x) - x

# Using method
root = secant_method(f, 0, 1, 1e-6)
print(f"Root: {root}")
```

1. The secant method does not require calculating the derivative, like the Newton-Raphson method.
2. The root of the function f(x)=cos(x)−x is successfully found between 0 and 1 with high accuracy.

5. Iteration Method

Example 1: Screenshot 2024-12-09 at 13-05-34 Сomputational Mathematics.png

```python
def fixed_point_iteration(g, x0, tol, max_iter):
    # Iterative root approximation
    x = x0
    for i in range(max_iter):
        x_next = g(x)
        if abs(x_next - x) < tol:  # Checking convergence
            return x_next
        x = x_next
    print("The method did not converge within the specified number of iterations.")
    return None

# Function transformation for iteration method
def g(x):
    return (x + 1)**(1/3)

# Using method
root = fixed_point_iteration(g, 1, 1e-6, 100)
print(f"Root: {root}")
```

1. The function f(x) is transformed to Screenshot 2024-12-09 at 13-07-19 Сomputational Mathematics.png, so as to find x=g(x).
2. The starting point x0​=1 is used for iterations.
3. The method is simple to implement, but may not converge if g′(x) is close to 1.

Example 2: f(x)=cos⁡(x)−x

```python
# Function transformation for iteration method
def g(x):
    return np.cos(x)

# Using method
root = fixed_point_iteration(g, 0.5, 1e-6, 100)
print(f"Root: {root}")
```

1. The equation f(x)=cos(x)−x is transformed into g(x)=cos(x).
2. Iterations continue until convergence or until the difference between iterations becomes less than tol.

6. Newton-Raphson Method for Complex Functions

Example: Screenshot 2024-12-09 at 13-11-18 Сomputational Mathematics.png

```python
def newton_raphson_complex(f, df, x0, tol):
    # Method
    x = x0
    while abs(f(x)) > tol:
        x = x - f(x) / df(x) #formule
    return x

# Definition of a function and its derivative
def f(x):
    return x * np.exp(x) - np.cos(x)

def df(x):
    return np.exp(x) * (x + 1) + np.sin(x)

# Using method
root = newton_raphson_complex(f, df, 0.5, 1e-6)
print(f"Root: {root}")
```

1. The Newton-Raphson method is well suited for equations involving exponential and trigonometric functions.
2. The derivative of the function f′(x) is calculated analytically to ensure fast convergence.

7. False Position Method

Example: Screenshot 2024-12-09 at 13-13-14 Сomputational Mathematics.png

```python
def false_position_method(f, a, b, tol):
    # We check that the root really lies in the interval [a, b]
    if f(a) * f(b) >= 0:
        print("Invalid initial values. f(a) and f(b) must be of different signs.")
        return None

    c = a 
    while abs(f(c)) > tol:
        c = b - f(b) * (b - a) / (f(b) - f(a))  # Formula
        if f(a) * f(c) < 0:
            b = c
        else:
            a = c
    return c

# Definition of a function
def f(x):
    return x**3 - 4*x - 9

# Using method
root = false_position_method(f, 2, 3, 1e-6)
print(f"Root: {root}")
```

1. The false position method is similar to the secant method, but additionally uses a sign change test to select the next interval.
2. The method is guaranteed to converge if the function is continuous and has a root in the given interval.

8. Muller's Method

Muller's Method is used to find both real and complex roots of nonlinear equations.

Example:Screenshot 2024-12-09 at 13-15-09 Сomputational Mathematics.png

```python
import cmath  # For working with complex numbers

def muller_method(f, x0, x1, x2, tol, max_iter=100):
    # Iterative root approximation by Muller's method
    for _ in range(max_iter):
        h0 = x1 - x0
        h1 = x2 - x1
        delta0 = (f(x1) - f(x0)) / h0
        delta1 = (f(x2) - f(x1)) / h1
        a = (delta1 - delta0) / (h1 + h0)
        b = a * h1 + delta1
        c = f(x2)

        discriminant = cmath.sqrt(b**2 - 4*a*c)
        if abs(b + discriminant) > abs(b - discriminant):
            denominator = b + discriminant
        else:
            denominator = b - discriminant

        x3 = x2 - (2 * c) / denominator  # Formula

        if abs(x3 - x2) < tol:  # Stop condition
            return x3
        x0, x1, x2 = x1, x2, x3

    print("The method did not converge within the specified number of iterations.")
    return None

# Definition of a function
def f(x):
    return x**3 - x**2 + 2

# Using method
root = muller_method(f, 0, 1, 2, 1e-6)
print(f"Root: {root}")
```

1. The method uses a parabola passing through three points x0​, x1​, x2​ to approximate the root.
2. Solving the quadratic equation gives the following approximation x3​.
3. The method is suitable for finding both real and complex roots.
